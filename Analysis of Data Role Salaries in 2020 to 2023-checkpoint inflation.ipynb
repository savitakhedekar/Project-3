```python
import matplotlib.pyplot as plt

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
ax.plot(DF['CountryCode'], DF['inflation_rate'], marker='o', linestyle='-', color='b', label='Value')

# Add labels and title
ax.set_xlabel('CountryCode')
ax.set_ylabel("inflation_rate")
ax.set_title('Plot of Value over Years')

# Add gridlines
ax.grid(True)

# Add legend
ax.legend()

# Show plot
plt.show()
```

```python
current_year = pd.Timestamp.now().year
start_year = current_year - 10
last_10_years_df = DF[(DF['year'] >= start_year) & (DF['year'] <= current_year)]
```

```python
grouped_data = last_10_years_df.groupby('CountryCode')['inflation_rate'].mean().reset_index()
```

```python
grouped_data
```

```python
plt.figure(figsize=(12, 8))
plt.bar(grouped_data['CountryCode'], grouped_data['inflation_rate'], color='skyblue')
plt.xlabel('CountryCode')
plt.ylabel('Average Inflation Rate')
plt.title('Average Inflation Rate Over the Last 10 Years by Country')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python

grouped_data = last_10_years_df.groupby('CountryCode')['inflation_rate'].mean().reset_index()
plt.figure(figsize=(10, 8))
plt.pie(grouped_data['inflation_rate'], labels=grouped_data['CountryCode'], autopct='%1.1f%%', startangle=140)
plt.title('Average Inflation Rate Distribution Over the Last 10 Years by Country')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
plt.figure(figsize=(14, 8))

# Step 4: Plot a line plot for inflation rate over time by country
for country in last_10_years_df['CountryCode'].unique():
    country_data = last_10_years_df[last_10_years_df['CountryCode'] == country]
    plt.plot(country_data['year'], country_data['inflation_rate'], marker='o', label=country)

# Adding titles and labels
plt.title('Inflation Rate Over the Last 10 Years by Country')
plt.xlabel('Year')
plt.ylabel('Inflation Rate')
plt.legend(title='CountryCode', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```
